Q.1 What is hypothesis testing in statistics?
Ans:- Hypothesis testing is a statistical technique used to test a hypothesis about a population parameter based on a sample of data. It involves formulating a hypothesis, collecting sample data, and using statistical methods to determine whether the hypothesis is supported or rejected.

The Hypothesis Testing Process-
* Null Hypothesis (H0): A statement of no effect or no difference. It is a default assumption that there is no significant difference or relationship.
* Alternative Hypothesis (H1 or Ha): A statement of an effect or difference. It is the hypothesis that there is a significant difference or relationship.
* Sample Data Collection: Collect a representative sample of data from the population.
* Test Statistic Calculation: Calculate a statistic that measures the difference between the sample and the null hypothesis.
* P-Value Calculation: Calculate the probability of observing the test statistic (or a more extreme value) assuming the null hypothesis is true.
* Decision Making: Compare the p-value to a predetermined significance level (usually 0.05). If the p-value is below the significance level, reject the null hypothesis. Otherwise, fail to reject the null hypothesis.
Long-Term Implications-
In the long term, hypothesis testing helps researchers and statisticians:
* Make Informed Decisions: By testing hypotheses, researchers can make informed decisions about populations based on sample data.
* Control for Errors: Hypothesis testing allows researchers to control for Type I and Type II errors, which helps to ensure the accuracy of conclusions.
* Identify Relationships: Hypothesis testing can help identify relationships between variables, which can inform future research and decision-making.
* Improve Predictive Models: By testing hypotheses about relationships between variables, researchers can improve predictive models and make more accurate predictions.
Common Applications-
Hypothesis testing has numerous applications in various fields, including:
* Medical Research: To test the effectiveness of new treatments or medications.
* Business: To test hypotheses about customer behavior, market trends, or product effectiveness.
* Social Sciences: To test hypotheses about social phenomena, such as the impact of education on income.
* Quality Control: To test hypotheses about product quality or manufacturing processes.

Q.2 What is the null hypothesis, and how does it differ from the alternative
hypothesis?
Ans:-Step 1: Understanding the Null Hypothesis
The null hypothesis (H0) is a fundamental concept in statistical hypothesis testing. It represents a default position that there is no effect or no difference; for instance, that a new treatment has no effect compared to the existing standard or that there's no correlation between two variables.

Step 2: Understanding the Alternative Hypothesis
The alternative hypothesis (H1 or Ha) is the statement that there is an effect or a difference. It represents the outcome that the study or experiment is designed to detect. The alternative hypothesis can be one-sided (directional) or two-sided (non-directional), depending on the research question.

Step 3: Long-term Implications of Null and Alternative Hypotheses
In the long term, the framework of null and alternative hypotheses allows researchers to systematically test theories and hypotheses. By assuming no effect initially (null hypothesis), researchers can use statistical tests to determine if the data provides sufficient evidence to reject this assumption in favor of the alternative hypothesis. This methodological approach helps in building a body of evidence over time, contributing to the advancement of knowledge in various fields.

Step 4: Decision Making Based on Hypotheses Testing
The decision to reject or not reject the null hypothesis is based on the p-value obtained from the statistical test. If the p-value is less than the chosen significance level (commonly set at 0.05), the null hypothesis is rejected in favor of the alternative hypothesis. This decision-making process, when consistently applied over many studies, helps in identifying true effects and differences, thereby guiding long-term strategies in research, medicine, business, and other areas.

Step 5: Importance in Research and Decision Making
The distinction between the null and alternative hypotheses is crucial for the integrity and validity of research findings. It ensures that conclusions are drawn based on statistical evidence rather than assumptions. Over the long term, this rigorous approach to hypothesis testing contributes to the reliability of research outcomes, informing policy, practice, and future research directions.

Q.3 Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
Ans:- Definition of Significance Level
The significance level, often denoted as alpha (α), is a threshold probability value used in hypothesis testing to determine whether the null hypothesis should be rejected. It represents the maximum probability of rejecting the null hypothesis when it is actually true (Type I error).

Step 1: Role in Hypothesis Testing
The significance level plays a crucial role in deciding the outcome of a hypothesis test. By setting a significance level before conducting the test, researchers can determine the critical region for rejecting the null hypothesis. If the p-value, which is the probability of observing the test results under the null hypothesis, is less than the significance level, the null hypothesis is rejected.

Step 2: Common Significance Levels
Commonly used significance levels include 0.05 and 0.01. A significance level of 0.05 means there is a 5% chance of rejecting the null hypothesis if it is true. Choosing a significance level depends on the context of the research and the consequences of Type I errors.

Step 3: Impact on Decision Making
The choice of significance level directly impacts the decision-making process in hypothesis testing. A lower significance level (e.g., 0.01) requires stronger evidence to reject the null hypothesis, reducing the chance of false positives but increasing the risk of false negatives (Type II errors). Conversely, a higher significance level (e.g., 0.10) makes it easier to detect effects but increases the risk of false positives.

Step 4: Conclusion
In conclusion, the significance level is a critical component of hypothesis testing that determines the threshold for rejecting the null hypothesis. It balances the risk of Type I errors against the need to detect true effects, influencing the outcome and interpretation of statistical tests.

Q.4 What are Type I and Type II errors? Give examples of each.
Ans:-Type I Error

- Definition: A Type I error occurs when a true null hypothesis (H0) is rejected in favor of an alternative hypothesis (H1), even though H0 is actually true.
- Also known as: False positive
- Example: A medical test indicates a person has a disease when they actually don't.
- Probability: The probability of making a Type I error is denoted by α (alpha) and is typically set to 0.05.

Type II Error

- Definition: A Type II error occurs when a false null hypothesis (H0) is not rejected in favor of an alternative hypothesis (H1), even though H1 is actually true.
- Also known as: False negative
- Example: A medical test fails to detect a disease in a person who actually has it.
- Probability: The probability of making a Type II error is denoted by β (beta).

Consequences

- Type I errors can lead to unnecessary actions or interventions.
- Type II errors can lead to missed opportunities or delayed interventions.

Control

- Type I errors can be controlled by setting a lower significance level (α).
- Type II errors can be controlled by increasing the sample size or improving the test's power.

Importance

- Understanding Type I and Type II errors is crucial in hypothesis testing and decision-making.
- It helps researchers and practitioners to consider the potential risks and consequences of their decisions.

Q.5 What is the difference between a Z-test and a T-test? Explain when to use
each.
Ans: Understanding Z-test:-
A Z-test is a statistical test used to determine whether a sample mean is significantly different from a known population mean. It assumes that the population standard deviation is known and the sample size is large (typically n > 30). The Z-test uses the standard normal distribution (Z-distribution) to calculate the test statistic.

 Understanding T-test:-To compare a sample mean to a known population mean or to compare the means of two groups
 Key differences:-
The key differences between Z-test and T-test are:
- Population standard deviation: Z-test requires the population standard deviation to be known, while T-test does not.
- Sample size: Z-test is typically used for large sample sizes (n > 30), while T-test is used for small sample sizes (n < 30).
- Distribution: Z-test uses the standard normal distribution, while T-test uses the t-distribution.
- Use Z-test:
    - When the population standard deviation is known.
    - When the sample size is large (n > 30).
    - For comparing a sample mean to a known population mean.
- Use T-test:
    - When the population standard deviation is unknown.
    -
    - For comparing the means of two groups or a sample mean to a known population mean.
In the choice between Z-test and T-test depends on the research question, sample size, and whether the population standard deviation is known. Understanding the differences between these tests is crucial for selecting the appropriate statistical method for hypothesis testing.

Q. Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.


```python
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 10
p = 0.5

binom_dist = binom.rvs(n, p, size=1000)
plt.hist(binom_dist, bins=11, alpha=0.7, align='left', rwidth=0.8)
plt.xlabel('Number of successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.show()
```

Q.7 Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

```python
import numpy as np
from scipy import stats

# Sample data
sample_data = np.array([49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9])

# Null hypothesis: population mean = 50
population_mean = 50
population_std = 1

# Calculate Z-statistic and p-value
z_statistic = (np.mean(sample_data) - population_mean) / (population_std / np.sqrt(len(sample_data)))
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print(f"Z-statistic: {z_statistic}, P-value: {p_value}")

if p_value < 0.05:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

Q.8 Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate data from a normal distribution
np.random.seed(0)
data = np.random.normal(loc=5, scale=2, size=100)

# Calculate the 95% confidence interval for the mean
mean = np.mean(data)
std_err = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=std_err)

print(f"Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")

# Plot the data
plt.hist(data, bins=20, alpha=0.7, label='Data')
plt.axvline(mean, color='r', label='Mean')
plt.axvline(ci[0], color='g', linestyle='--', label='95% CI')
plt.axvline(ci[1], color='g', linestyle='--')
plt.legend()
plt.show()
```

Q.9 Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(5, 2, 1000)
z_scores = (data - np.mean(data)) / np.std(data)

plt.hist(data, bins=20, alpha=0.5, label='Original Data')
plt.hist(z_scores, bins=20, alpha=0.5, label='Z-scores')
plt.legend()
plt.show()
```
